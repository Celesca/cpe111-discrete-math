## Permutations

__Permutation__ is the act of arranging all the members of a set into some sequence or order, a process called __permuting__.

To find the number of $k$-permutations of $A$, by calculating $\frac{|A|!}{(|A|-k)!}$. <br> 
* the special case of $k=|A|$, which is equivalent to finding the number of ways of ordering the elements of $A$. <br>

Python uses the  **itertools** library for permuations and combinations.

```python
from itertools import permutations
```

## The Job-Assignment Problem
* How to put the right person on the right job
* $n$ people who need to be assigned to execute $n$ jobs (1 person: 1 job)
* The cost (time taking or wage) of the $i^{th}$ person when assigned to the $j^{th}$ job is in a table $C$ , a $C[i,j]$ for each pair $i,j = 1,2,..,n$

### Compare Job-Assignment problem's solution between 2 approaches : Greedy and Brute-force

### Method 1 - Based on Greedy Algorithm

```python
def greedy_job_assignment(cost_matrix):
    num_workers, num_tasks = cost_matrix.shape
    assignment = [-1] * num_workers  # Initialize with no assignments

    for worker in range(num_workers):
        min_cost = float('inf') # Infinite
        min_task = -1
        # find min cost for each task along each person
        for task in range(num_tasks):
            if task not in assignment:  # Check if the task is unassigned
                if cost_matrix[worker][task] < min_cost: # Compare to select the min hour that worker can do.
                    min_cost = cost_matrix[worker][task]
                    min_task = task

        assignment[worker] = min_task  # Assign the worker to the task

    return assignment

assignment = greedy_job_assignment(cost_matrix)
total_time = 0
print("With Greedy Assignment:")
for i, task in enumerate(assignment):
    print("{0:10} is assigned to task: {1}".format(Persons[i],Tasks[task]))
    total_time += cost_matrix[i,task]
print("Total time: {0:3d} hours".format(total_time))
```

## Knapsack problem 

#### (เช่น Lazada จัดของยังไงให้น้ำหนักและมูลค่าและเร็ว) มันคือ ปัญหา Optimize

* Given $n$ itmes of know weight $w_1, w_2,...,w_n$ and values $v_1, v_2,...,v_n$ and a knapsack of capacity $W$ <br>


### find the most value subset of the items that fit into the knapsack

```python
items = [(60,10), (100,20), (120,30), (50, 25)] #(value, weight)
knapsack_capacity = 50
```

### `Practice4` <br>

### Compare the Knapsack problem's solution between 2 approaches: Greedy Algorithm and Brute-force Algorithm

### Method 1 -  Based-on Greedy Algorithm
#### จริงๆ Greedy เราเอาเป็นทึ่มๆ เช็คน้ำหนักแล้วใส่เรียงตามลำดับได้เลย

```python
def greedy_knapsack(items, capacity):
    # Calculate value:weight ratio for each item
    value_per_weight = [(item[0] / item[1], item[0], item[1]) for item in items]
    # Sort items by value-to-weight ratio in descending order
    value_per_weight.sort(reverse=True) #เรียงให้มันมีมูลค่ามากที่สุดก่อนในอัตราส่วน
    total_value = 0.0
    total_weight = 0.0
    knapsack = []
    for ratio, value, weight in value_per_weight: # Multiple variables
        if (total_weight + weight) <= capacity:
            # Add the whole item to the knapsack
            knapsack.append((value, weight))
            total_value += value
            total_weight += weight
        else:
            break
    return knapsack, total_value, total_weight 
```

```python
selected_items, total_value, total_weight = greedy_knapsack(items, knapsack_capacity)
print("Knapsack capacity:", knapsack_capacity)
print("List of items (value,weight):",items)
print("--------------------------------------")
print("Total Value:", total_value)
print("Total Weight", total_weight)
print("Selected Items:", selected_items)
print("--------------------------------------")
```
