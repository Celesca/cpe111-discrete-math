### Boolean expressions and Boolean Functions


Definition: Let $B = {0, 1}$. Then $B^n$ = ${(x_1, x_2, …, x_n) | x_i ∈ B for 1 ≤ i ≤ n }$ is the set of all possible $n$-tuples of 0s and 1s.   
The variable $x$ is called a Boolean variable if it assumes values only from $B$, that is, if its only possible values are 0 and 1.   
A function from $B^n to B$ is called a Boolean function of degree $n$.

__Example 1:__ The function $F(x, y) = x\bar{y}$ from the set of ordered pairs of Boolean variables to the set {0, 1} is a Boolean function of degree 2

```python
from tabulate import tabulate

def boolean_function(x, y):
    return int(bool(x * (not y )))

# Generate the truth table
truth_table = []
for x in [0, 1]:  # 0 or 1
    for y in [0, 1]:
        result = boolean_function(x, y)
        truth_table.append([x, y, int(bool(result))]) #cast int to boolean and cast to int again

# Display the truth table using tabulate
headers = ["x", "y", "F(x,y) = x.~y"]
table = tabulate(truth_table, headers=headers, tablefmt="simple_grid")
print(table)
```

__Example 2:__ Find the values of the Boolean function represented by $F(x, y, z) = xy + \bar z$

```python
from tabulate import tabulate

def boolean_function(x, y, z):
    return int(x and y) + int(not z)

# Generate the truth table
truth_table = []
for x in [0, 1]:  # 0 or 1
    for y in [0, 1]:
        for z in [0, 1]:
            result = boolean_function(x, y, z)
            truth_table.append([x, y, z, int(bool(result))]) #cast int to boolean and cast to int again

# Display the truth table using tabulate
headers = ["x", "y", "z", "F(x,y,z) = xy + ~z"]
table = tabulate(truth_table, headers=headers, tablefmt="simple_grid")
print(table)
```

### __Summary: Identities of Boolean Algebra__
| Identity| Name|
|---|---|
|$\bar{\bar{x}} = x$ | Law of the double complement|
|$x + x = x$ <br> $x.x = x$| Idempotent laws| 
|$x + 0 = x$ <br> $x.1 = x$| Identity laws|
|$x + 1 = 1$ <br> $x.0 = 0$| Domination laws|
|$x + y = y + x$ <br> $x.y = y.x$| Commutation laws|

more
| Identity| Name|
|---|---|
|$x+(y+z) = (x+y)+z$ <br> $x(yz) = (xy)z$| Associative laws|
|$x + yz = (x+y)(x+z)$ <br> $x(y+z) = xy + xz$| Distributive laws|
|$\bar{(xy)} = \bar{x} + \bar{y}$| De Morgan's laws|
|$x+xy=x$ <br> $x(x+y)=x$| Absorption laws|
|$x+\bar{x} = 1$| Unit property|
|$x\bar{x} = 0$| Zero property|


All identities, except for the first and the last two come in pairs. Each element of the pair is `the dual` of the other (obtained by switching Boolean sums and Boolean products and 0’s and 1’s).




### Duality
`The dual` of a Boolean expression is obtained by interchanging Boolean sums and Boolean products and interchanging 0s and 1s

```python
# code here
from tabulate import tabulate

def boolean_function1(x, y, z):
    return x and int(y or z)

def boolean_function2(x,y,z):
    return int(x and y) or int(x and z)

def result(x,y,z):
    return int(bool(boolean_function1(x,y,z))) == int(bool(boolean_function2(x,y,z)))

# Generate the truth table
truth_table = []
for x in [0, 1]:  # 0 or 1
    for y in [0, 1]:
        for z in [0, 1]:
            truth_table.append([x, y, z, y or z, x and y , x and z , int(bool(boolean_function1(x,y,z))) , int(bool(boolean_function2(x,y,z))) , int(bool(result(x,y,z)))]) #cast int to boolean and cast to int again

# Display the truth table using tabulate
headers = ["x", "y", "z", "y+z" , "xy" , "xz" , "x(y+z)" , "xy + xz" , "yes/no"]
table = tabulate(truth_table, headers=headers, tablefmt="simple_grid")
print(table)
```

### Representing Boolean functions (ผลบวกของฟังก์ชัน ก็คือเขาให้ผลลัพธ์มา เราก็แค่หา Boolean Expression)
#### Sum-of-Products Expansions
* `A literal` is a Boolean variable or its complement.
* `A minterm` of the Boolean variables $x_1, x_2,..., x_n$ is a Boolean product $y_1, y_2, ...y_n$ where $y_i = x_i$ or $y_i = \bar{x_i}$ Hence, a minterm is a product of $n$ literals, with one literal for each variable
<br>
* The minterm $y_1, y_2,..., y_n$ has value 1 if and only if each $x_i$ is 1.This occurs if and only if $x_i = 1$ when
$y_i = x_i$ and $x_i = 0$ when $y_i = \bar{x_i}$
<br>
* __The sum of minterms__ that represents the function is called `the sum-of-products expansion` or __the disjunctive normal form of the Boolean function__
