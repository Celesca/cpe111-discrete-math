````python
#Example: An urn contains four blue balls and five red balls.
#What is the probability that a ball chosen from the urn is blue?
URN = ["Blue", "Blue", "Blue", "Blue", "Red", "Red", "Red", "Red", "Red"]
choose_Ball_all = list(combinations(URN,1))
print(choose_Ball_all)
print("All case: {0}".format(len(choose_Ball_all)))
choose_Blue_Ball = [ball for ball in choose_Ball_all if ball == ("Blue",) ]
print("happen case: {0}".format(len(choose_Blue_Ball)))
print("Probability that a blue ball is chosen from the urn: {0:.2f}".format(len(choose_Blue_Ball)/ len(choose_Ball_all)))
````

## Examples of Distribution

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Uniform distribution
low = 0
high = 1
size = 1000
random_samples = np.random.uniform(low,high,size)
# Plotting
# ให้ Plot กราฟเดียวมีสองรูป โดยมี 1 แถว 2 Columns
fig, axes = plt.subplots(1,2, figsize=(12,4))
ax1 = axes[0] # พิกัดของกราฟรูปแรก และรูปสอง
ax2 = axes[1]
ax1.plot(random_samples)
ax1.set_title("Uniform Samples")
ax2.hist(random_samples,bins = 10, edgecolor="black") # Histogram bins = (จำนวนช่วง , แท่ง)
ax2.set_title("Uniform Distribution")
plt.show()
````

````python
# Normal distribution
# Set a seed for reproducibility (optional)
np.random.seed(42)
# Generate random samples from a normal distribution
mean = 0
std_dev = 1
size = 1000
random_samples = np.random.normal(mean, std_dev, size)
# Plotting
fig, axes = plt.subplots(1,2, figsize=(12,4))
ax1 = axes[0]
ax2 = axes[1]
ax1.plot(random_samples)
ax1.set_title("Normal distrubution Samples")

# Plotting the histogram
ax2.hist(random_samples, bins=10, density=True, alpha=0.7, edgecolor='black')
# Plotting the probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean)/std_dev)**2)
ax2.plot(x, pdf, 'k--', linewidth=2)
# Adding labels and title
ax2.set_title('Normal Distribution')
ax2.set(xlabel='Value', ylabel='Probability Density')

plt.show()
````

__Random seed__ is a starting point used in generating __pseudorandom numbers__. <br>
In computer programming, randomization is not truly random but is __Pseudorandom__ that the random numbers are generated using deterministic algorithms.  <br>

&nbsp;&nbsp;&nbsp;These algorithms produce sequences of numbers that appear to be random, __but__ it's not. it is the same random-like sequence. <br>

&nbsp;&nbsp;&nbsp;So if we know the starting point (the seed), we can guess the next numbers either reproduce the entire sequence.

Setting a random seed is useful in scenarios where you want to ensure reproducibility. By using the same seed, you get the same sequence of pseudorandom numbers every time you run the program. This is particularly important in scientific research, simulations, and other fields where reproducibility is crucial for verification and debugging.

## Python's random module V.S. Numpy's random module

**Remark: Refer to answers from ChatGPT3.5** <br>
Both the standard Python `random` module and the NumPy `random` module provide functionality for generating random numbers, but they have some differences in terms of features and use cases.



### Python `random` module:

1. **Generality:** The `random` module in Python is part of the Python Standard Library and is a general-purpose module for generating pseudorandom numbers. It provides functions for generating random integers, floats, and making random choices.

2. **Basic Random Functions:**
   ```python
   import random

   # Random float in the range [0.0, 1.0)
   random_float = random.random()

   # Random integer in the range [a, b]
   random_integer = random.randint(a, b)

   # Random choice from a sequence
   random_choice = random.choice(sequence)
   ```

3. **Random Seed:**
   ```python
   random.seed(seed_value)
   ```
   Allows you to set a seed for reproducibility.

### NumPy `random` module:

1. **Array Operations:** The `numpy.random` module is part of the NumPy library and is more focused on generating random numbers for numerical operations, especially for arrays.

2. **Array-Based Random Functions:**
   ```python
   import numpy as np

   # Random float in the range [0.0, 1.0)
   random_float = np.random.rand()

   # Random array of floats in the range [0.0, 1.0)
   random_array = np.random.rand(5, 5)

   # Random integer from a discrete uniform distribution
   random_integer = np.random.randint(a, b, size=(5, 5))
   ```

3. **Advanced Distributions:**
   ```python
   # Random samples from a normal distribution
   random_normal = np.random.normal(mean, std_dev, size=(5, 5))
   ```

4. **Seed in a Different Way:**
   ```python
   np.random.seed(seed_value)
   ```
   Sets the seed for the NumPy random number generator.

### Summary:

- If you need simple random number generation for general purposes, the standard `random` module in Python is sufficient.
  
- If you are working with numerical operations and arrays, especially in the context of data science or scientific computing, NumPy's `random` module provides more advanced functionality and is often preferred.

Both modules have their use cases, and the choice depends on the requirements of your specific task.

````python
# Example of correlation of sine waves
import numpy as np
import matplotlib.pyplot as plt

# Create a time array
t = np.linspace(0, 2 * np.pi, 1000)

# Define two sine waves with different phase shifts
sin_wave_1 = np.sin(t)
sin_wave_2 = np.sin(t + np.pi/4)  # Phase shift of pi/4

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(sin_wave_1, sin_wave_2)[0, 1]

# Plot the sine waves
plt.plot(t, sin_wave_1, label='sin(x)')
plt.plot(t, sin_wave_2, label='sin( x + $\pi / 4 $ )')
plt.title(f'Correlation Coefficient: {correlation_coefficient:.2f}')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
````

### **Finding a Mode:**

````python
import numpy as np

def find_mode(arr):
    unique_values, counts = np.unique(arr, return_counts=True)
    max_count_index = np.argmax(counts)
    mode_value = unique_values[max_count_index]
    return mode_value
````

````python
# Example array
data = np.random.normal(50,10,100)
data_int = data.astype(int)
print(data_int)
# Find mode using the custom function
mode_value = find_mode(data_int)

# Print the mode
print("Mode:", mode_value)
````

## Expected value and Varience

ค่าเฉลี่ย Expected Value สูตรคือ ตัวมันเองคูณกับโอกาสที่มันจะเกิดขึ้น

### Explected value
The expected value is a key concept in probability theory and statistics. It represents the average or mean value of a random variable based on its probability distribution. Mathematically, the expected value $E(X)$ of a discrete random variable $X$ is calculated as the sum of each possible value of $X$ multiplied by its probability: <br>

$$E(X)= \sum_{i} x_i \cdot P(X=x_i)$$

The variance ($Var[X]$) of a random variable ($X$) measures how much the values of ($X$) vary on average from the expected value ($E[X]$). The variance is calculated using the formula:

$$Var[X] = E[(X - E[X])^2]$$

In this formula:
- $E[X]$ is the expected value of the random variable $X$.
- $Var[X]$ is the variance of $X$.
- $X$ represents each individual value of the random variable.

````python
import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

# Create a series of random values using numpy.random.choice
# In this example
# let's create a series of 1000 values, choosing from the numbers 1, 2, and 3 with probability = [0.3, 0.5, 0.2].
p = [0.3, 0.5, 0.2] # โอกาสการเกิด (weight)
data = np.random.choice([1, 2, 3],p=p, size=1000)

# Calculate the expected value (mean) of the series
expected_value = np.mean(data)
variance_value = np.var(data)

# Print the results
print("Expected Value:", expected_value)
print("Variance Value:", variance_value)
````
